```python
# Librerias necesarias
import pandas as pd
import numpy as np

# Visualizacion
import matplotlib.pyplot as plt
import seaborn as sns

# Reglas de asociacion
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Cargar dataset
df = pd.read_csv("https://raw.githubusercontent.com/MendezBy02/Parcial4-Abel-Antonio-Escobar-Mendez-25-1449-2022/refs/heads/main/archivos/clave_B_asociacion.csv")

# Mostrar primeras filas
df.head()
```

Explicacion:
Se carga el archivo CSV que contiene las transacciones de compra de los clientes.
Luego se muestran las primeras filas para entender la estructura del dataset.

```python
# Informacion general
df.info()

# Verificar nulos
df.isnull().sum()

# Verificar duplicados
df.duplicated().sum()
```

Explicacion:
Se revisan tipos de datos, valores nulos y registros duplicados para asegurar la calidad de los datos antes del analisis.

Preparar datos para Apriori

```python
df.head()
```

```python
# Agrupar productos por transaccion
transacciones = df.groupby('transaccion_id')['item'].apply(list).tolist()

# Mostrar primeras transacciones
transacciones[:5]
```

Explicacion:
Se agrupan los productos por cada transaccion para obtener listas de productos comprados juntos.
Este formato es necesario para aplicar Apriori.

```python
# Convertir transacciones a formato binario
te = TransactionEncoder()

te_array = te.fit(transacciones).transform(transacciones)

basket = pd.DataFrame(te_array, columns=te.columns_)

# Mostrar primeras filas
basket.head()
```

Explicacion:
Se transforma el dataset en una matriz binaria donde cada producto tiene valores True o False dependiendo si fue comprado o no en cada transaccion.

```python
# Aplicar algoritmo Apriori
frequent_itemsets = apriori(
    basket,
    min_support=0.02,
    use_colnames=True
)

# Mostrar resultados
frequent_itemsets.head()
```

Explicacion:
El algoritmo Apriori identifica combinaciones frecuentes de productos utilizando soporte minimo.

```python
# Generar reglas de asociacion
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.3
)

# Mostrar reglas
rules.head()
```

```python
# Top 10 reglas mas relevantes
top_rules = rules.sort_values(by='lift', ascending=False)

top_rules[['antecedents',
           'consequents',
           'support',
           'confidence',
           'lift']].head(10)
```

Explicacion:
Se ordenan las reglas segun el lift,
permitiendo identificar las relaciones
mas fuertes entre productos.
